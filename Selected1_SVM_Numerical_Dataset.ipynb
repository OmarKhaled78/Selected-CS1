```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('BreastCancerDataset.csv')
```

```python
dataset #Preview Dataset
```

```python
#Data Preprocessing
# 1- dropping the ID column because it should not affect the model's prediction
dataset.drop('id' , axis = 1 , inplace = True)
```

```python
dataset # dataset preview after dropping ID column
```

```python
#2-Dropping last column as it contains all NaN values
dataset.drop(dataset.columns[31] , axis = 1 ,inplace = True)
dataset #dataset preview after dropping column with NaN values
```

```python
#Features
X = dataset.iloc[: , 1:].values
print (X)
```

```python
Y = dataset.iloc[: , 0].values
print(Y)
```

```python
#splitting dataset into training set and testing set
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
#Feature scaling using standardscaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test)
```

```python
#feature preview after standardization
print(X_train) #training features
```

```python
print(X_test) #testing features
```

```python
#training
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf' , random_state=0)
classifier.fit(X_train , Y_train)
```

```python
#predicting
Y_prediction = classifier.predict(X_test)
print(Y_prediction)
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test , Y_prediction))
```

```python
#confusion matrix 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
cm = confusion_matrix(Y_test,Y_prediction)
print(cm)
cm_visualisation = ConfusionMatrixDisplay(cm).plot()
```

```python
#ROC Curve
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay
Y_prediction2 = classifier.decision_function(X_test)
fpr,tpr , threshold = roc_curve(Y_test , Y_prediction2 , pos_label = classifier.classes_[1])
roc_curvedisplay = RocCurveDisplay(fpr = fpr , tpr = tpr).plot()
```
